```python
import tensorflow as tf 
import numpy as np
import matplotlib.pyplot as plt
```

Defino los datos de entrada y de salida en este caso son los grados celsius y los grados fahrenheit respectivamente.

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

Creo la capa de salida diciendo que voy a tener una neurona de entrada y una neurona de salida.
Tambien para crear el modelo de red neuronal uso keras (Keras es una API de redes neuronales de alto nivel, escrita en Python y capaz de ejecutarse sobre TensorFlow, CNTK o Theano. Fue desarrollada con el objetivo de permitir una experimentación rápida. Te crea sola las redes neuronales y te las entrena.)
Sequential es un tipo de modelo de keras que es una pila lineal de capas de red neuronal. Es para modelos simples de red neuronal.

```python
# capa = tf.keras.layers.Dense(units=1, input_shape=[1])
# modelo = tf.keras.Sequential([capa])

oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)
modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

Para compilar el modelo antes de entrenarlo utilizamos el otimizador Adam que es el que ayuda a no perder tiempo y siempre acercarnos al resultado correcto. y para el modelo del error usamos el error cuadratico medio.

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

Lo entrenamos al modelo con la función fit dando la entrada y salida, epochs es la cantidad de veces que se va a entrenar el modelo, batch_size es la cantidad de muestras que se van a usar para calcular el error y verbose es para que muestre el progreso del entrenamiento.

```python
print("Comenzando entrenamiento...")
historial = modelo.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print("Modelo entrenado!")
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de perdida")
plt.plot(historial.history["loss"])
```

```python
print("Hagamos una predicción!")
resultado = modelo.predict([100.0])
print("El resultado es " + str(resultado) + " fahrenheit!")
```
